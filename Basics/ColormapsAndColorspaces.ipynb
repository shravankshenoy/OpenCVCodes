<h3> ColorMaps </h3>

Note that a gray scale image is shown as a coloured image. This is because img is just a 2D matrix in which there are no colors, only a "luminosity" value. So matplotlib applies a colormap by default. Therfore, if we want to print a grayscale image, we must specify colormap as gray

https://stackoverflow.com/questions/41265576/plt-imshow-shows-color-images-for-grayscale-images-in-ipython

```python
plt.imshow(img, cmap="gray")
plt.xticks([]) # to remove labelling on x axis
plt.show()
```

```python
plt.imshow(img, cmap="ocean")
plt.yticks([]) # remove labelling on y axis

plt.show()
```

<h3> ColorSpaces </h3>

A color space is a specific organization of colors. Color space, is an abstract mathematical model which simply describes the range of colors as tuples of numbers, typically as 3 or 4 values or color components. 

The various ways in which we represent color channels are known as color spaces. Some common color spaces are(decompose color of image into specific components):

RGB

BGR

HSV - Hue Saturation Value

CMY -

Note : In RGB color space, if you increase magnitude of R of a pixel, in result of that not only redness will increase, brightness of that pixel will also increase. In HSV, color and brightness are independent. HSV color space is more near to human perception.

https://www.quora.com/Why-do-we-use-the-HSV-color-space-so-often-in-vision-and-image-processing



```python
img = cv2.imread(img_path , 1)
```

```python
plt.imshow(img)
plt.show()
```

The color of the image is distorted. This is because cv2 reads the image as BGR but matplotlib plots the image assuming the image is in RGB colorspace. To remove the distortion, we have to convert the image to that colorspace which is suitable for matplotlib

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show()
```
