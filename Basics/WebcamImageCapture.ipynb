```python
import cv2
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture(0) # 0 implies first webcam
    
if cap.isOpened(): # check if camera running successfully
    # ret tells whether image was captured successfully or not
    # frame contains the image
    ret, frame = cap.read()
    print(ret)
    print(frame)
else:
    ret = False

img1 = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
 
    
plt.imshow(img1)
plt.title('Color Image RGB')
plt.xticks([])
plt.yticks([])
plt.show()
    

cap.release()
```
