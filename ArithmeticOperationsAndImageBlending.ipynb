<h3> Arithmetic Operations </h3>

```python
path = "C:\\Users\\Dell\\Documents\\ImageProcessing\\"
    
imgpath1 =  path + "airplane.png"
imgpath2 =  path + "tulips.png"

img1 = cv2.imread(imgpath1, 1)
img2 = cv2.imread(imgpath2, 1)

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img2 = img2[:,:512,:]
```

```python
add = img1 + img2
diff = img1- img2
mult = img1 * 2
div = img2 / 3
```

```python
images = [img1,img2, add,diff,mult,div]
```

```python
plt.figure(figsize=(10,8))
for i in range(6):
    plt.subplot(1,6,i+1)
    plt.imshow(images[i])
```

<h3> Image Blending </h3>

```python
alpha = 0.5 #Weight to first image
beta = 0.5 #Weight to second image
gamma = 0

```

```python
output = cv2.addWeighted(img1,alpha,img2,beta,gamma)
```

```python
plt.imshow(output)
```
